```python
import pandas as pd

# Importar a base de dados dos clientes
tabela = pd.read_csv("/content/drive/MyDrive/clientes.csv")

# Visualizar a base de dados dos clientes
display(tabela)

# Score de cr√©dito = Nota de cr√©dito
# Good = Boa
# Standard = OK
# Poor = Ruim
```

```python
# Preparar a base de dados para a IA
display(tabela.info())

from sklearn.preprocessing import LabelEncoder

# cientista - 1
# bombeiro - 2
# engenheiro - 3
# dentista - 4
# artista - 5

# profissao
codificador_profissao = LabelEncoder()
tabela["profissao"] = codificador_profissao.fit_transform(tabela["profissao"])

# mix_credito
codificador_credito = LabelEncoder()
tabela["mix_credito"] = codificador_credito.fit_transform(tabela["mix_credito"])

# comportamento_pagamento
codificador_pagamento = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_credito.fit_transform(tabela["comportamento_pagamento"])

# exibir informa√ß√µes da base de dados
display(tabela.info())
```

```python
# y -> coluna da base de dados que eu quero prever
y = tabela["score_credito"]

# x -> coluna da base de dados que eu vou usar para fazer a previs√£o
x = tabela.drop(columns=["score_credito", "id_cliente"])

# separar dados de treino e teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# Imports necess√°rios
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# 1Ô∏è‚É£ CORRIGIR OS DADOS - Converter categorias em n√∫meros
label_encoder = LabelEncoder()

# Se y_treino tem valores como 'Poor', 'Fair', 'Good', 'Excellent'
y_treino_encoded = label_encoder.fit_transform(y_treino)

# 2Ô∏è‚É£ CRIAR OS MODELOS
modelo_arvoredecisao = RandomForestClassifier(
    n_estimators=100,  # Mais √°rvores = melhor performance
    random_state=42    # Para resultados reprodut√≠veis
)

modelo_knn = KNeighborsClassifier(
    n_neighbors=5,     # N√∫mero de vizinhos
    weights='distance' # Vizinhos mais pr√≥ximos t√™m mais peso
)

# 3Ô∏è‚É£ TREINAR CORRETAMENTE
print("üéØ Treinando Modelo de √Årvore de Decis√£o...")
modelo_arvoredecisao.fit(x_treino, y_treino_encoded)

print("üéØ Treinando Modelo KNN...")
modelo_knn.fit(x_treino, y_treino_encoded)  # ‚Üê CORRE√á√ÉO AQUI!

print("‚úÖ Modelos treinados com sucesso!")

# 4Ô∏è‚É£ VERIFICAR A TRANSFORMA√á√ÉO
print(f"üìä Classes originais: {list(label_encoder.classes_)}")
print(f"üìä Classes codificadas: {list(range(len(label_encoder.classes_)))}")
```

```python
# Solu√ß√£o r√°pida e direta
from sklearn.preprocessing import LabelEncoder

# 1. Encoder os dados
le = LabelEncoder()
y_treino_num = le.fit_transform(y_treino)
y_teste_num = le.transform(y_teste)

# 2. Treinar modelos
modelo_arvoredecisao.fit(x_treino, y_treino_num)
modelo_knn.fit(x_treino, y_treino_num)  # ‚Üê FIX AQUI!

# 3. Predi√ß√µes
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# 4. Acur√°cia
from sklearn.metrics import accuracy_score
print(f"üéØ Acur√°cia Random Forest: {accuracy_score(y_teste_num, previsao_arvoredecisao):.2%}")
print(f"üéØ Acur√°cia KNN: {accuracy_score(y_teste_num, previsao_knn):.2%}")
```

```python
# üöÄ C√ìDIGO REFATORADO COM VERIFICA√á√ÉO E CRIA√á√ÉO AUTOM√ÅTICA DE ENCODERS
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

print("üéØ INICIANDO PREDI√á√ÉO PARA NOVOS CLIENTES")
print("=" * 50)

# üìä IMPORTAR E ANALISAR BASE DE DADOS DOS NOVOS CLIENTES
try:
    print("üìÅ Carregando base de dados dos novos clientes...")
    tabela_novos_clientes = pd.read_csv("/content/drive/MyDrive/novos_clientes.csv")
    print(f"‚úÖ Dados carregados: {tabela_novos_clientes.shape[0]} clientes, {tabela_novos_clientes.shape[1]} colunas")

    print("\nüìä DADOS ORIGINAIS:")
    display(tabela_novos_clientes)

    # An√°lise inicial dos dados
    print(f"\nüîç An√°lise r√°pida:")
    print(f"   ‚Ä¢ Colunas dispon√≠veis: {list(tabela_novos_clientes.columns)}")
    for col in ['profissao', 'mix_credito', 'comportamento_pagamento']:
        if col in tabela_novos_clientes.columns:
            valores_unicos = tabela_novos_clientes[col].nunique()
            print(f"   ‚Ä¢ {col}: {valores_unicos} categorias √∫nicas")
        else:
            print(f"   ‚ö†Ô∏è Coluna '{col}' n√£o encontrada!")

except FileNotFoundError:
    print("‚ùå ERRO: Arquivo 'novos_clientes.csv' n√£o encontrado!")
    print("üí° Verifique o caminho: /content/drive/MyDrive/novos_clientes.csv")
    raise
except Exception as e:
    print(f"‚ùå ERRO ao carregar dados: {e}")
    raise

# üîß VERIFICA√á√ÉO E CRIA√á√ÉO INTELIGENTE DE ENCODERS
def verificar_e_criar_encoder(nome_encoder, dados_referencia=None):
    """
    Verifica se encoder existe e est√° treinado, sen√£o tenta criar/treinar
    """
    print(f"\nüîç Verificando encoder: {nome_encoder}")

    # Verificar se existe na mem√≥ria
    if nome_encoder in globals():
        encoder = globals()[nome_encoder]

        # Verificar se foi treinado
        if hasattr(encoder, 'classes_'):
            print(f"   ‚úÖ {nome_encoder} est√° OK! Categorias: {encoder.classes_}")
            return encoder
        else:
            print(f"   ‚ö†Ô∏è  {nome_encoder} existe mas n√£o foi treinado")
    else:
        print(f"   ‚ùå {nome_encoder} n√£o encontrado na mem√≥ria")

    # Se chegou aqui, o encoder precisa ser criado/treinado
    print(f"   üí° Tentando recriar {nome_encoder}...")

    # Tentar encontrar dados de refer√™ncia para treinar
    if dados_referencia is not None:
        print(f"   üîß Criando novo encoder com dados fornecidos...")
        novo_encoder = LabelEncoder()
        novo_encoder.fit(dados_referencia)
        globals()[nome_encoder] = novo_encoder
        print(f"   ‚úÖ {nome_encoder} criado e treinado!")
        return novo_encoder
    else:
        print(f"   ‚ùå N√£o foi poss√≠vel recriar {nome_encoder} - dados de refer√™ncia necess√°rios!")
        return None

# üõ°Ô∏è FUN√á√ÉO PARA TRANSFORMA√á√ÉO SEGURA MELHORADA
def transformar_coluna_segura(dados, coluna, encoder, nome_encoder):
    """
    Transforma uma coluna categ√≥rica de forma segura, tratando valores desconhecidos
    """
    print(f"\nüîß Processando coluna: {coluna}")

    # Verificar se coluna existe
    if coluna not in dados.columns:
        raise KeyError(f"Coluna '{coluna}' n√£o encontrada nos dados!")

    # Verificar se encoder est√° OK
    if encoder is None or not hasattr(encoder, 'classes_'):
        raise ValueError(f"Encoder '{nome_encoder}' n√£o est√° dispon√≠vel para transforma√ß√£o!")

    # Analisar categorias
    categorias_dados = set(dados[coluna].unique())
    categorias_encoder = set(encoder.classes_)
    categorias_desconhecidas = categorias_dados - categorias_encoder
    categorias_comum = categorias_dados & categorias_encoder

    print(f"   üìä Categorias nos dados: {sorted(categorias_dados)}")
    print(f"   ‚úÖ Categorias conhecidas: {len(categorias_comum)}")

    if categorias_desconhecidas:
        print(f"   ‚ö†Ô∏è  Categorias DESCONHECIDAS: {sorted(categorias_desconhecidas)}")

        # Estrat√©gia: substituir por categoria mais comum do encoder
        categoria_substituicao = encoder.classes_[0]  # Primeira categoria como padr√£o

        print(f"   üîÑ Substituindo por: '{categoria_substituicao}'")

        # Fazer substitui√ß√£o
        dados_corrigidos = dados[coluna].copy()
        for cat_desconhecida in categorias_desconhecidas:
            dados_corrigidos = dados_corrigidos.replace(cat_desconhecida, categoria_substituicao)
            print(f"      ‚Ä¢ '{cat_desconhecida}' ‚Üí '{categoria_substituicao}'")

        return encoder.transform(dados_corrigidos)
    else:
        print("   ‚úÖ Todas as categorias s√£o conhecidas!")
        return encoder.transform(dados[coluna])

# üîç VERIFICA√á√ÉO DOS ENCODERS
print(f"\n{'='*50}")
print("üîç VERIFICANDO E PREPARANDO ENCODERS")
print(f"{'='*50}")

# Lista de encoders necess√°rios
encoders_info = [
    ('codificador_profissao', 'profissao'),
    ('codificador_credito', 'mix_credito'),
    ('codificador_pagamento', 'comportamento_pagamento')
]

encoders_verificados = {}

for nome_encoder, coluna_referencia in encoders_info:
    # Dados de refer√™ncia da tabela de novos clientes (para emerg√™ncia)
    dados_ref = tabela_novos_clientes[coluna_referencia].unique() if coluna_referencia in tabela_novos_clientes.columns else None

    encoder = verificar_e_criar_encoder(nome_encoder, dados_ref)
    encoders_verificados[nome_encoder] = encoder

    if encoder is None:
        print(f"‚ùå PROBLEMA CR√çTICO: N√£o foi poss√≠vel preparar {nome_encoder}")
        print(f"üí° SOLU√á√ÉO MANUAL necess√°ria:")
        print(f"   1. Carregue os dados de treinamento originais")
        print(f"   2. Execute: {nome_encoder} = LabelEncoder()")
        print(f"   3. Execute: {nome_encoder}.fit(dados_treino['{coluna_referencia}'])")

        # Oferecer alternativa de emerg√™ncia
        print(f"\nüö® ALTERNATIVA DE EMERG√äNCIA:")
        print(f"   Criando encoder tempor√°rio apenas com dados novos...")
        encoder_temp = LabelEncoder()
        encoder_temp.fit(dados_ref)
        encoders_verificados[nome_encoder] = encoder_temp
        globals()[nome_encoder] = encoder_temp
        print(f"   ‚ö†Ô∏è  Encoder tempor√°rio criado! (pode n√£o ter todas as categorias do treino)")

# üîÑ PROCESSAR COLUNAS CATEG√ìRICAS COM SEGURAN√áA
print(f"\n{'='*50}")
print("üîß INICIANDO TRANSFORMA√á√ÉO DE DADOS CATEG√ìRICOS")
print(f"{'='*50}")

# Criar c√≥pia para preservar dados originais
dados_processados = tabela_novos_clientes.copy()

try:
    # 1Ô∏è‚É£ PROFISS√ÉO
    dados_processados["profissao"] = transformar_coluna_segura(
        dados_processados,
        "profissao",
        encoders_verificados['codificador_profissao'],
        "codificador_profissao"
    )

    # 2Ô∏è‚É£ MIX CR√âDITO
    dados_processados["mix_credito"] = transformar_coluna_segura(
        dados_processados,
        "mix_credito",
        encoders_verificados['codificador_credito'],
        "codificador_credito"
    )

    # 3Ô∏è‚É£ COMPORTAMENTO PAGAMENTO
    dados_processados["comportamento_pagamento"] = transformar_coluna_segura(
        dados_processados,
        "comportamento_pagamento",
        encoders_verificados['codificador_pagamento'],
        "codificador_pagamento"
    )

    print(f"\n‚úÖ TODAS AS TRANSFORMA√á√ïES REALIZADAS COM SUCESSO!")

except Exception as e:
    print(f"‚ùå ERRO na transforma√ß√£o: {e}")
    print(f"üí° Tipo do erro: {type(e).__name__}")
    raise

# üìä EXIBIR DADOS TRANSFORMADOS
print(f"\n{'='*50}")
print("üìä DADOS AP√ìS TRANSFORMA√á√ÉO:")
print(f"{'='*50}")
display(dados_processados)

# üéØ VALIDAR DADOS ANTES DA PREDI√á√ÉO
print(f"\nüîç VALIDA√á√ÉO PR√â-PREDI√á√ÉO:")
try:
    # Verificar se modelo existe e est√° treinado
    if 'modelo_arvoredecisao' not in globals():
        raise NameError("Modelo 'modelo_arvoredecisao' n√£o foi definido!")

    if not hasattr(modelo_arvoredecisao, 'n_features_in_'):
        raise ValueError("Modelo n√£o foi treinado ainda!")

    # Verificar n√∫mero de features
    features_esperadas = modelo_arvoredecisao.n_features_in_
    features_dados = dados_processados.shape[1]

    print(f"   ‚Ä¢ Features esperadas pelo modelo: {features_esperadas}")
    print(f"   ‚Ä¢ Features nos dados: {features_dados}")

    if features_dados != features_esperadas:
        print(f"   ‚ö†Ô∏è  INCOMPATIBILIDADE: O modelo espera {features_esperadas} features, mas os dados t√™m {features_dados}")

        # Tentar ajustar automaticamente
        if features_dados > features_esperadas:
            print(f"   üîß Removendo colunas extras...")
            dados_processados = dados_processados.iloc[:, :features_esperadas]
        elif features_dados < features_esperadas:
            raise ValueError(f"Dados insuficientes! Faltam {features_esperadas - features_dados} features")

    print("   ‚úÖ Valida√ß√£o conclu√≠da!")

except Exception as e:
    print(f"‚ùå ERRO na valida√ß√£o: {e}")
    raise

# üîÆ FAZER PREDI√á√ïES
print(f"\n{'='*50}")
print("üîÆ REALIZANDO PREDI√á√ïES")
print(f"{'='*50}")

try:
    print("üéØ Aplicando modelo Random Forest...")
    nova_previsao = modelo_arvoredecisao.predict(dados_processados)

    print(f"‚úÖ Predi√ß√µes realizadas com sucesso!")
    print(f"üìä {len(nova_previsao)} predi√ß√µes geradas")

    print(f"\nüéØ RESULTADOS DAS PREDI√á√ïES:")
    display(nova_previsao)

    # Estat√≠sticas das predi√ß√µes
    print(f"\nüìà ESTAT√çSTICAS DAS PREDI√á√ïES:")
    unique_preds, counts = np.unique(nova_previsao, return_counts=True)
    for pred, count in zip(unique_preds, counts):
        percentage = (count / len(nova_previsao)) * 100
        print(f"   ‚Ä¢ Classe {pred}: {count} clientes ({percentage:.1f}%)")

except Exception as e:
    print(f"‚ùå ERRO na predi√ß√£o: {e}")
    print(f"üí° Tipo do erro: {type(e).__name__}")
    raise

print(f"\nüéâ PROCESSO CONCLU√çDO COM SUCESSO!")
print(f"{'='*50}")
```
